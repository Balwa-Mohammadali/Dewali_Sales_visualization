```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('Diwali Sales Data.csv',encoding='latin')

# print(df.head())

df=df.dropna(subset=['Gender','Age Group','Age','State'])

gender=df['Gender'].value_counts()
plt.bar(gender.index,gender.values,color=['pink','yellow'])
plt.title('gender radio ')
plt.xlabel('gender')
plt.ylabel('no. of male/female')
plt.tight_layout()
plt.show()

age_group=df['Age Group'].value_counts()
plt.pie(age_group.values,labels=age_group.index,autopct='%1.1f%%',colors=['green','grey','lightblue','pink','yellow','red','orange'])
plt.title('pie chart of age group ')
plt.tight_layout()
plt.show()


age=df['Age']
plt.hist(age,bins=10,edgecolor='black',color='purple')
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('no. of customber')
plt.tight_layout()
plt.show()

plt.figure(figsize=(22,4))
state=df['State'].value_counts()
plt.scatter(state.index,state.values,color='red',label='state customber')
plt.title('state distribution')
plt.xlabel('state')
plt.ylabel('no. of costumber')
plt.legend()
plt.tight_layout()
plt.show()
```
