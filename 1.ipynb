```python
#line graph
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,20,15,30]
plt.plot(x,y)
plt.show()
```

```python
#line graph
import matplotlib.pyplot as plt

x=['mon','tue','wed','thr','fri']
y=[10,5,20,11,25]

plt.plot(x,y,color='red',linestyle='--',linewidth='2',marker='o',label='2025 bakery sales')
plt.title('bakery sales')
plt.xlabel('day of week')
plt.ylabel('sales per day')
plt.legend(loc='lower right',fontsize='12')
plt.grid(color='black',linewidth='2',linestyle=':')
plt.xlim('mon','fri')
plt.ylim(0,25)
plt.yticks([5,10,11,20,25],['5pro','10pro','11pro','20pro','25pro'])
plt.show()
```

```python
#bar graph
import matplotlib.pyplot as plt
product=[1,2,3,4]
sales=[10,40,35,20]
plt.bar(product,sales,color='orange',width=.5 ,label='2025 analysis of product selling')
plt.title('product sale comparision')
plt.xlabel('product')
plt.ylabel('sales of product')
plt.legend(loc='lower right')
plt.xlim(0,5)
plt.ylim(0,45)
plt.xticks([1,2,3,4],['pant','shirt','t-shirt','jeans'])

plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt

region=['north','south','east','west']
revenue=[3500,1000,2000,1500]

plt.pie(revenue,labels=region,autopct='%1.1f%%',colors=['pink','orange','skyblue','lightgreen'])
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt
score=[45,66,76,97,88,65,79,81,55,49,39,86,92,77,83,71,63]
plt.hist(score,bins=5,color='darkgreen',edgecolor='black')
plt.xlabel('score of students')
plt.ylabel('no. of students')
plt.title('students score distribution')
plt.show()
```

```python
#scatter graph
import matplotlib.pyplot as plt
hour_time=[1,2,3,4,5]
marks=[45,50,55,60,65]

plt.scatter(hour_time,marks,marker='s',color='blue',label='students data')
#marker='D-diamond,^-triangle,o-circle,*-cross,s-square etc...'
plt.xlabel('hours students study')
plt.ylabel('marks of students')
plt.title('relation between time and marks')
plt.grid()
plt.legend()
plt.show()
```

```python
#scatter graph
import matplotlib.pyplot as plt

plt.scatter([1,2,3,4,5],[45,50,55,60,65],marker='^',color='blue',label='class A')
plt.scatter([1,2,3,4,5],[40,45,50,58,67],marker='s',color='orange',label='class B')
plt.xlabel('hours students study')
plt.ylabel('marks of students')
plt.title('relation between class A and B')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt


x=[1,2,3,4]
y=[10,20,15,30]

plt.subplot(1,2,1)
plt.plot(x,y,color='darkblue')
plt.title('line graph')


plt.subplot(1,2,2)
plt.bar(x,y,color='green')
plt.title('bar graph')

plt.suptitle('comaparision of line and bar')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots(1,2,figsize=(10,5))


x=[1,2,3,4]
y=[10,20,15,30]

ax[0].plot(x,y,color='darkblue')
ax[0].set_title('line graph')

ax[1].bar(x,y,color='green')
ax[1].set_title('bar graph')

plt.suptitle('comaparision of line and bar')

plt.tight_layout()
plt.savefig('graph.png',dpi=100,bbox_inches='tight')#dpi-300 for high resolution
plt.show()
```
